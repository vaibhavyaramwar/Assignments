```python
def useYield():
    for i in range(0,10000):
        yield i+2
        print(i)
```

```python
for i in useYield():
    print(i)
```

```python
def genCube(n):
    for i in range(n):
         yield i**3
```

```python
for i in genCube(100000):
    print(i)
```

```python
def fibo(n):
    a,b=1,1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
for i in fibo(100):
    print(i)
```

```python
def fact(n):
    a=1
    for i in range(1,n):
        a=a*i
        yield a
```

```python
for i in fact(5):
    print(i)
```
